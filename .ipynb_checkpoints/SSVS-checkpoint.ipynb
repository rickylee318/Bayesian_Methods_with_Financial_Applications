```python
N=100
K=100

#Data simulation
X = np.random.uniform(low = 0, high = 4, size = (N,K-1))
X = np.concatenate((np.ones((N,1)),X), axis=1)
beta_0_true = -2
beta_1_true = 1
beta_2_true = 2
beta_3_true = 3

sigma = 1 # known
y = np.random.normal(beta_0_true * X[:,0]  + beta_1_true * X[:,1] + beta_2_true * X[:,2] + beta_3_true * X[:,3], sigma)
```

```python
#hyperparameter
c_0, d_0 = 1., 1.
tau, c = 0.5, 8.

#initial value
beta = np.ones((K,))
W = st.beta.rvs(c_0, d_0, size=K)
gamma = np.array([])
for w in W:
    gamma = np.append(gamma, bernoulli.rvs(w))

#posterior
pos_beta = []
pos_W = []
pos_gamma = []

for i in range(10000):
    d = gamma * c * tau + (1 - gamma) * tau
    D = np.diag(d)

    beta_variance = sigma**2 * np.linalg.inv(np.dot(X.T,X) + np.linalg.inv(D)**2 * sigma **2)
    beta_mu = np.dot(X.T,y) / sigma**2
    beta_mu = np.dot(beta_variance, beta_mu)
    beta = np.random.multivariate_normal(beta_mu.flatten(), beta_variance)

    W = st.beta.rvs(c_0+gamma, 1+i+d_0-gamma) #i

    a = norm.pdf(beta, 0, c*tau) * W
    b = norm.pdf(beta, 0, tau) * W
    gamma = bernoulli.rvs(a/(a+b))

    pos_beta.append(beta)
    pos_W.append(W)
    pos_gamma.append(gamma)
```

```python
pos_beta = pd.DataFrame(pos_beta)
pos_beta.mean()
```

```python
%matplotlib inline
plt.figure(figsize=(30, 20))
plt.style.use('ggplot')
plt.plot(pos_beta.iloc[:,0], label = '0')
plt.plot(pos_beta.iloc[:,1], label = '1')
plt.plot(pos_beta.iloc[:,2], label = '2')
plt.plot(pos_beta.iloc[:,3], label = '3')
plt.plot(pos_beta.iloc[:,4], label = '4')
plt.legend(fontsize=20)
plt.show()
```
