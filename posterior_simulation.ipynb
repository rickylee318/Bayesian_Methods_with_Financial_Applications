```python
import math
import random
import numpy as np
import pandas as pd
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
import scipy.stats as st
%matplotlib inline
# %precision 4
```

```python
sigma = 1 #known

#true value
beta_1_true = 1

# data simulation
N = 50
# N = [5,10,100,500]

# for n in N:
x = np.random.uniform(low = -4, high = 4, size = N)
y = np.random.normal(beta_1_true * x, sigma)

#prior 
var_prior = 0.01
beta_prior = 0

#likelihood
likelihood =[]
r = np.linspace(-2, 2, 500)
for rr in r:
    likelihood.append(st.norm.pdf(y,rr * x,sigma).prod())
likelihood = likelihood/sum(likelihood)

#posterior
beta_post = (sum(x*y)/sigma**2) / (sum(x**2)/sigma**2 + 1 / var_prior)
var_post = 1 / (sum(x**2)/sigma**2 + 1 / var_prior)


plt.figure(figsize=(12, 9))
plt.style.use('ggplot')
plt.plot(r, st.norm.pdf(r,beta_prior,np.sqrt(var_prior))/sum(st.norm.pdf(r,beta_prior,np.sqrt(var_prior))),label='prior',c='blue')
plt.plot(r, likelihood,label='likelihood',c='green')
plt.plot(r, st.norm.pdf(r,beta_post,np.sqrt(var_post))/sum(st.norm.pdf(r,beta_post,np.sqrt(var_post))),label='posterior',c='red')
plt.ylabel('Density', fontsize=16)
plt.legend()
# plt.show()
```
