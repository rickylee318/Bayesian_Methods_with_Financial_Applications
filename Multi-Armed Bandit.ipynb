# Multi-Armed Bandit Introducation and Application

###  Assume we have two arms each with an associated reward probability 

```python
random.seed(3)
```

```python
class Bandit(object):
    def __init__(self, num_options=2, prior=(1.0, 1.0)):
        self.trials = np.zeros(shape=(num_options,), dtype=int)
        self.successes = np.zeros(shape=(num_options,), dtype=int)
        self.num_options = num_options
        self.prior = prior
        self.alphas = []
        self.betas = []

    def update(self, choice, conv):
        self.trials[choice] = self.trials[choice] + 1
        if (conv):
            self.successes[choice] = self.successes[choice] + 1

    def evaluate(self):
        sampled_theta = []
        alpha = np.zeros(shape=(self.num_options,), dtype=int)
        bet = np.zeros(shape=(self.num_options,), dtype=int)
        for i in range(self.num_options):
            alpha[i] = self.prior[0] + self.successes[i]
            bet[i] = self.prior[1] + self.trials[i] - self.successes[i]
            dist = beta(alpha[i],bet[i])
            sampled_theta += [dist.rvs()]
        self.alphas.append(alpha)
        self.betas.append(bet)
        return sampled_theta.index(max(sampled_theta))
```

```python
theta = (0.3, 0.4)

def click(c):
    if random.random() < theta[c]:
        return True
    else:
        return False
```

```python
N = 10000
incr_trials = np.zeros(shape=(N, 2))
successes = np.zeros(shape=(N, 2))

b = Bandit()
for i in range(N):
    choice = b.evaluate()
    conv = click(choice)
    b.update(choice, conv)
    incr_trials[i] = b.trials
    successes[i] = b.successes
```

```python
first_hyper_alphas = []
first_hyper_betas = []
second_hyper_alphas = []
second_hyper_betas = []
legend = ['1','1000','2000','3000','4000','5000','6000','7000','8000','9000','10000']
for i in range(0,10000,999):
    first_hyper_alphas.append(b.alphas[i][0])
    first_hyper_betas.append(b.betas[i][0])
    second_hyper_alphas.append(b.alphas[i][1])
    second_hyper_betas.append(b.betas[i][1])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
for a,b,l in zip(first_hyper_alphas, first_hyper_betas, legend):
    dist = beta(a, b)
    x = np.linspace(0, 1, 1000)[1:-1]
    plt.plot(x, dist.pdf(x), label = l, linewidth=2)

plt.legend(loc = 'best', fontsize=20)
plt.title('first posterior of beta', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
for a,b,l in zip(second_hyper_alphas, second_hyper_betas, legend):
    dist = beta(a, b)
    x = np.linspace(0, 1, 1000)[1:-1]
    plt.plot(x, dist.pdf(x), label = l, linewidth=2)

plt.legend(loc = 'best', fontsize=20)
plt.title('second posterior of beta', fontsize=20)
plt.show()
```

```python
symbol_list = ['JPM', 'TGT', 'WFC', 'AAPL', 'WMT', 'COST', 'GS', 'AMZN', 'MSFT']
start = datetime(2002, 6, 1)
end = date.today()
reward = pd.DataFrame()
sym_key_val = []
for i, sym in enumerate(symbol_list):
    symbol = symbol_db.dataReader(sym, start, end, 'D', 'db')
    re = symbol['Adj Close'] - symbol['Adj Open']
    sym_key_val.append((sym, i))
    reward = pd.concat((reward, re), axis=1)
reward.columns = symbol_list
trading_days, num_stocks = reward.shape
```

```python
class Bandit(object):
    def __init__(self, num_options=num_stocks, prior=(1.0, 1.0)):
        self.trials = np.zeros(shape=(num_options,), dtype=int)
        self.successes = np.zeros(shape=(num_options,), dtype=int)
        self.num_options = num_options
        self.reward = []
        self.prior = prior
        self.alphas = []
        self.betas = []
        self.sym = []

    def update(self, choice, conv, t):
        self.trials[choice] = self.trials[choice] + 1
        c = sym_key_val[choice][0]
        self.reward.append(reward[c].iloc[t])
        self.sym.append(c)
        if (conv):
            self.successes[choice] = self.successes[choice] + 1

    def evaluate(self):
        sampled_theta = []
        alpha = np.zeros(shape=(self.num_options,), dtype=int)
        bet = np.zeros(shape=(self.num_options,), dtype=int)
        for i in range(self.num_options):
            alpha[i] = self.prior[0] + self.successes[i]
            bet[i] = self.prior[1] + self.trials[i] - self.successes[i]
            dist = beta(alpha[i],bet[i])
            sampled_theta += [dist.rvs()]
        self.alphas.append(alpha)
        self.betas.append(bet)
        return np.argmax(sampled_theta)
```

```python
def click(c, t):
    choice = sym_key_val[c][0]
    if reward[choice].iloc[t] > 0 :
        return True
    else:
        return False
```
